## Uniform Distribution

```python
from scipy.stats import uniform
```

```python
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc=start, scale=width)
```

```python
data_uniform
```

```python
ax = sns.distplot(data_uniform,
                 bins=100,
                 kde=True,
                 color='skyblue',
                 hist_kws={'linewidth':15,
                          'alpha':1})
ax.set(xlabel='Uniform Distribution',ylabel='Frequency')
```

## Confidence Intervals

```python
import pandas as pd
import numpy as np
df = pd.read_csv('Heart.csv')
df.head()
```

```python
df['sex1'] = df.sex.replace({1:'Male',
                            0: 'Female'})
```

```python
dx = df[['target', 'sex1']].dropna()
```

```python
pd.crosstab(dx.target, dx.sex1)
```

```python
p_fm = 226/(86+226)
p_fm
```

```python
n = 86+226
se_female = np.sqrt(p_fm * (1 - p_fm) / n)
se_female
```

```python
z_score = 1.96
lcb = p_fm - z_score * se_female
ucb = p_fm + z_score * se_female

lcb, ucb
```

```python
import statsmodels.api as sm
sm.stats.proportion_confint(n * p_fm, n)
```

```python
# STEP 2 - POPULATION PROPORTION
p_us = len(df[df['target']==1])/len(df)
p_us
```

```python
# STEP 3 - CALCULATE THE TEST STATISTICS
se = np.sqrt(0.42 * (1-0.42) / len(df))
se
```

```python
be = p_us
he = 0.42
test_stat = (be - he) / se
test_stat
```

```python
# STEP 4 - CALCULATE THE P-VALUE
pvalue = 2*dist.norm.cdf(-np.abs(test_stat))
pvalue
```
